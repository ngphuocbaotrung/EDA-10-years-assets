```python
asset_name = 'ETF'
df = price_data[[asset_name]].reset_index()
df.columns = ['ds', 'y']
```

```python
forecast_years = 5
inflation_rate = 0.035
discount_factor = (1 + inflation_rate) ** forecast_years
future_returns = {}

# Tạo biểu đồ nhiều subplot
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 12))
axes = axes.flatten()

assets = ['Bitcoin', 'Gold', 'ETF', 'Stock']

for i, asset in enumerate(assets):
    df = price_data[[asset]].reset_index()
    df.columns = ['ds', 'y']

    model = Prophet()
    model.fit(df)

    future = model.make_future_dataframe(periods=5*365)
    forecast = model.predict(future)

    # Vẽ biểu đồ dự báo
    model.plot(forecast, ax=axes[i])
    axes[i].set_title(f"{asset} Forecast", fontsize=14)
    axes[i].set_xlabel("Date")
    axes[i].set_ylabel("Portfolio Value (€)")

    # Tính % tăng trưởng & giá trị thực
    initial_value = df['y'].iloc[-1]
    final_nominal = forecast['yhat'].iloc[-1]
    real_value = final_nominal / discount_factor
    growth_pct = (final_nominal / initial_value - 1) * 100

    # Gắn chú thích trên biểu đồ
    axes[i].annotate(f"+{growth_pct:.1f}%",
                     xy=(forecast['ds'].iloc[-1], final_nominal),
                     xytext=(-100, 20),
                     textcoords='offset points',
                     arrowprops=dict(arrowstyle='->', color='green'),
                     fontsize=10, color='green')

    axes[i].annotate(f"Real (inflation-adjusted): €{real_value:,.0f}",
                     xy=(forecast['ds'].iloc[-1], final_nominal),
                     xytext=(-100, -20),
                     textcoords='offset points',
                     fontsize=9, color='gray')

    # Lưu kết quả
    future_returns[asset] = {
        'Forecast (€)': round(final_nominal, 2),
        'Real Forecast (€)': round(real_value, 2),
        'Growth (%)': round(growth_pct, 2)
    }

# Xoá subplot thừa
if len(assets) < len(axes):
    for j in range(len(assets), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# Xuất bảng tổng hợp kết quả
future_df = pd.DataFrame(future_returns).T
future_df['Growth (%)'] = future_df['Growth (%)'].map(lambda x: f"{x:.2f}%")
future_df['Forecast (€)'] = future_df['Forecast (€)'].map(lambda x: f"€{x:,.2f}")
future_df['Real Forecast (€)'] = future_df['Real Forecast (€)'].map(lambda x: f"€{x:,.2f}")
print(future_df)
```

```python
# Step 1: Collect detailed forecast data
detailed_forecasts = []

for asset in assets:
    df = price_data[[asset]].reset_index()
    df.columns = ['ds', 'y']

    model = Prophet()
    model.fit(df)

    future = model.make_future_dataframe(periods=5*365)
    forecast = model.predict(future)

    # Calculate real (inflation-adjusted) forecast
    forecast['Asset'] = asset
    forecast['Nominal Forecast (€)'] = forecast['yhat']
    forecast['Real Forecast (€)'] = forecast['yhat'] / discount_factor
    forecast = forecast[['ds', 'Asset', 'Nominal Forecast (€)', 'Real Forecast (€)']]

    # Keep only future dates
    forecast = forecast[forecast['ds'] > df['ds'].max()]

    detailed_forecasts.append(forecast)

# Step 2: Combine all assets into one DataFrame
all_forecast_df = pd.concat(detailed_forecasts, ignore_index=True)

# Optional: Format values for better readability
all_forecast_df['Nominal Forecast (€)'] = all_forecast_df['Nominal Forecast (€)'].map(lambda x: round(x, 2))
all_forecast_df['Real Forecast (€)'] = all_forecast_df['Real Forecast (€)'].map(lambda x: round(x, 2))

# Step 3: Display or export
print(all_forecast_df.head(20))

# Optional: Export to CSV
# all_forecast_df.to_csv("5_year_detailed_forecast.csv", index=False)
```

```python
# Step 1: Filter from today to 5 years later
today = pd.to_datetime("today").normalize()
future_df = forecast_df[(forecast_df['ds'] >= today) & (forecast_df['ds'] <= today + pd.Timedelta(days=5*365))].copy()

# Step 2: Simulate 1€/day investment per asset
results = {}

for asset in future_df['Asset'].unique():
    df = future_df[future_df['Asset'] == asset].copy()
    df = df.dropna(subset=['Nominal Forecast (€)'])  # ensure no NaN
    df['Units Bought'] = 1 / df['Nominal Forecast (€)']
    total_units = df['Units Bought'].sum()
    final_price = df['Nominal Forecast (€)'].iloc[-1]
    
    total_value = total_units * final_price
    total_invested = len(df)
    
    results[asset] = {
        "Total Invested (€)": total_invested,
        "Final Units": round(total_units, 4),
        "Final Price (€)": round(final_price, 2),
        "Portfolio Value (€)": round(total_value, 2),
        "Profit (€)": round(total_value - total_invested, 2),
        "ROI (%)": round((total_value / total_invested - 1) * 100, 2)
    }

# Step 3: Display the result
roi_df = pd.DataFrame(results).T.sort_values("ROI (%)", ascending=False)
roi_df
```
